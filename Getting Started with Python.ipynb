#  Loops
Python has two primitive loop commands:

-  while loops : With the while loop we can execute a set of statements as long as a condition is true.
-  for loops : A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

##  While loops
Require an indexing variable to set beforehand.

```python
#Print i as long as i is less than 6:

i = 1
while i < 6:
  print(i)
  i += 1        # does the same as i=i+1, increasing the current value of i by 1
```

The `break` and `continue` Statement:

-  With the break statement we can stop the loop even if the while condition is true.
-  With the continue statement we can stop the current iteration, and continue with the next.

```python
# Exit the loop when i is 3:

i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
#Continue to the next iteration if i is 3:

i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

##  For loops 
Does not require an indexing variable to set beforehand.

```python
# Print each fruit in a fruit list:

fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
# Using the range() function:

for x in range(6):
  print(x)
```

```python
# Increment the sequence with 3 (start, stop, increment):

for x in range(2, 30, 3):
  print(x)
```

```python
# Increment the sequence with 3 for decimal numbers (start, stop, length):

for x in np.linspace(2, 30, 10):
  print(x)
```

##  Slicing
Slicing in Python refers to the process of extracting a portion of a sequence (such as a list, tuple, or string) by specifying a range of indices. It allows you to create a new sequence containing only the elements within the specified range.

The basic syntax for slicing is: **sequence[start_index:end_index:step]** <br>

-  `start_index`: The index from which the slicing starts (inclusive). If omitted, slicing starts from the beginning of the sequence.
-  `end_index`: The index where the slicing stops (exclusive). If omitted, slicing continues to the end of the sequence.
-  `step`: Optional. Specifies the increment between indices. If omitted, the default value is 1.

```python
# Slicing a list
my_list = [1, 2, 3, 4, 5]
print(my_list[1:4])     # Output: [2, 3, 4]
print(my_list[:3])      # Output: [1, 2, 3]
print(my_list[2:])      # Output: [3, 4, 5]
print(my_list[::2])     # Output: [1, 3, 5]  #selects every second element starting from the first one

# Slicing a string
my_string = "Hello, World!"
print(my_string[7:])    # Output: World!
print(my_string[:5])    # Output: Hello
print(my_string[::2])   # Output: Hlo ol!

# Slicing with negative indices
print(my_list[:-1])     # Output: [1, 2, 3, 4]
print(my_string[:-1])   # Output: Hello, World
```

Seaborn's `lineplot()` function performs aggregation by default, meaning that if multiple data points share the same x-coordinate, Seaborn will aggregate them (usually by calculating the mean) and connect the aggregated points with lines. Matplotlib's plot() function does not perform aggregation by default and will connect the data points directly with lines.

The code `plt.plot(wcgs3['sdp'], wcgs3['weight'])` resulting in too many lines instead of one line suggests that the data might not be sorted by the x-values, causing the plot to connect points in an unexpected way.

```python
plt.figure(figsize=(8, 4))
######################################################
##### Create a boxplot using Matplotlib
data = [wcgs2['weight'][wcgs2['arcus'] == category] for category in wcgs2['arcus'].unique()]

plt.subplot(1, 2, 1)
plt.boxplot(data, labels=wcgs2['arcus'].unique(), notch=True, widths=0.8)
plt.title('Boxplot (Matplotlib)')
plt.xlabel('arcus')
plt.ylabel('weight')
#######################################################

##### Create a boxplot using Seaborn
plt.subplot(1, 2, 2)
sns.boxplot(data=wcgs2, x="arcus", y="weight", notch=True)
plt.title('Boxplot (Seaborn)')
########################################################

plt.show()
```

```python
plt.figure(figsize=(8, 4))
######################################################
##### Create a Histogram using Matplotlib
plt.subplot(1, 2, 1)
plt.hist(wcgs2['weight'], bins=30, color='skyblue', edgecolor='black', density=True) 
plt.title('Histogram (Matplotlib)')
plt.xlabel('weight')
plt.ylabel('Count')

# Overlay KDE plot
sns.kdeplot(wcgs2['weight'], color='cornflowerblue', linewidth=2)  # Use Seaborn's kdeplot for KDE plot
#######################################################

##### Create a Histogram using Seaborn
plt.subplot(1, 2, 2)
sns.histplot(data=wcgs2, x="weight", kde=True, bins=30)
plt.title('Histogram (Seaborn)')
########################################################

plt.show()
```

**Correlation Plot:**

```python
# Calculate the correlation matrix
corr = wcgs3.corr()


plt.figure(figsize=(10, 5))

# Create a heatmap with matplotlib
plt.subplot(1, 2, 1)
heatmap = plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar(heatmap)
plt.title('Correlation Heatmap (Matplotlib)') # x and y axis NOT automaticall labelled.

# Set ticks and tick labels for x and y axes
plt.xticks(ticks=range(len(corr.columns)), labels=corr.columns, rotation=90)
plt.yticks(ticks=range(len(corr.columns)), labels=corr.columns)


# Create a heatmap with seaborn
plt.subplot(1, 2, 2)
sns.heatmap(corr, annot=False, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap (Seaborn)') # x and y axis automaticall labelled.

plt.tight_layout()
plt.show()
```
